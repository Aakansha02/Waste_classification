```python
import torch
import torch.nn as nn
from efficientnet_pytorch import EfficientNet
from torch.utils.data import DataLoader
from torchvision import transforms
from torchvision.datasets import ImageFolder  # <-- Import ImageFolder here
from torch.optim import Adam
import matplotlib.pyplot as plt
```

### Preprocessing (Transforms)
#### EfficientNet requires input images to be resized to a fixed size. Typically, EfficientNet models are trained with images of size 224x224, but it might vary based on the specific EfficientNet version. need to normalize the images according to the EfficientNets pre-trained weights:

```python
transform = transforms.Compose([
    transforms.Resize((224, 224)),        # Resize images to 224x224
    transforms.ToTensor(),                # Convert image to tensor
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),  # Normalize with ImageNet stats
    transforms.RandomHorizontalFlip(),    # Augmentation: Random flip
    transforms.RandomRotation(30),        # Augmentation: Random rotation
])
```

```python
train_data = ImageFolder(root=r"D:\waste classification\DATASET\TRAIN", transform=transform)
val_data = ImageFolder(root=r"D:\waste classification\DATASET\TEST", transform=transform)

train_loader = DataLoader(train_data, batch_size=32, shuffle=True)
val_loader = DataLoader(val_data, batch_size=32, shuffle=False)
```

```python
# Load the EfficientNet model (B0 is a smaller version, you can try B7 for more performance)
model = EfficientNet.from_pretrained('efficientnet-b0')

# Modify the final fully connected layer to output 3 classes (organic, recyclable, non-recyclable)
model._fc = nn.Linear(model._fc.in_features, 3)  # 3 classes for your dataset
```

```python
criterion = nn.CrossEntropyLoss()  # Suitable for multi-class classification
optimizer = Adam(model.parameters(), lr=1e-4)  # You can experiment with learning rates
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

num_epochs = 10  # Adjust as needed

for epoch in range(num_epochs):
    model.train()  # Set model to training mode
    running_loss = 0.0
    correct = 0
    total = 0

    for inputs, labels in train_loader:
        inputs, labels = inputs.to(device), labels.to(device)

        # Zero the parameter gradients
        optimizer.zero_grad()

        # Forward pass
        outputs = model(inputs)
        loss = criterion(outputs, labels)

        # Backward pass and optimization
        loss.backward()
        optimizer.step()

        # Calculate accuracy
        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

        running_loss += loss.item()

    epoch_loss = running_loss / len(train_loader)
    epoch_accuracy = 100 * correct / total
    print(f'Epoch [{epoch + 1}/{num_epochs}], Loss: {epoch_loss:.4f}, Accuracy: {epoch_accuracy:.2f}%')

    # Validation step (to monitor overfitting)
    model.eval()  # Set model to evaluation mode
    with torch.no_grad():
        val_correct = 0
        val_total = 0
        for inputs, labels in val_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            _, predicted = torch.max(outputs, 1)
            val_total += labels.size(0)
            val_correct += (predicted == labels).sum().item()

        val_accuracy = 100 * val_correct / val_total
        print(f'Validation Accuracy: {val_accuracy:.2f}%')

```

### validation

```python
# Validation step
model.eval()  # Set model to evaluation mode
with torch.no_grad():  # Disable gradient computation for validation
    val_correct = 0
    val_total = 0
    for inputs, labels in val_loader:
        inputs, labels = inputs.to(device), labels.to(device)
        outputs = model(inputs)
        _, predicted = torch.max(outputs, 1)
        val_total += labels.size(0)
        val_correct += (predicted == labels).sum().item()

    val_accuracy = 100 * val_correct / val_total
    print(f'Validation Accuracy: {val_accuracy:.2f}%')
```
